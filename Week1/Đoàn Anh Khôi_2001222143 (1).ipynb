```

# 1. Trộn 2 mảng 1 chiều
def tron_mang_min(a, b):
    c = []
    min_len = min(len(a), len(b))

    for i in range(min_len):
        c.append(min(a[i], b[i]))

    if len(a) > min_len:
        c.extend(a[min_len:])
    if len(b) > min_len:
        c.extend(b[min_len:])

    return c

a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]
print("Mảng kết quả:", tron_mang_min(a, b))
```

```

# 2. Các hàm xử lý ma trận
def tong_tam_giac_tren_cheo_phu(a):
    n = len(a)
    tong = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:
                tong += a[i][j]
    return tong

def doi_am_thanh_duong(a):
    return [[abs(x) for x in row] for row in a]

def thay_chan_bang_x(a, x):
    return [[x if val % 2 == 0 else val for val in row] for row in a]

def toan_chan(a):
    return all(val % 2 == 0 for row in a for val in row)

def doi_xung(a):
    n = len(a)
    for i in range(n):
        for j in range(n):
            if a[i][j] != a[j][i]:
                return False
    return True

def cheo_chinh_tang(a):
    n = len(a)
    for i in range(n - 1):
        if a[i][i] >= a[i + 1][i + 1]:
            return False
    return True

def tam_giac_duoi_cheo_phu(a):
    n = len(a)
    return [a[i][j] for i in range(n) for j in range(n) if i + j >= n - 1]

def cheo_phu_giam(a):
    n = len(a)
    for i in range(n - 1):
        if a[i][n - 1 - i] <= a[i + 1][n - 2 - i]:
            return False
    return True

mt = [
    [1, -2, 3],
    [4, 5, -6],
    [7, 8, 9]
]

print("Tổng tam giác trên chéo phụ:", tong_tam_giac_tren_cheo_phu(mt))
print("Đổi âm:", doi_am_thanh_duong(mt))
print("Thay chẵn bằng 0:", thay_chan_bang_x(mt, 0))
print("Toàn chẵn:", toan_chan(mt))
print("Đối xứng:", doi_xung(mt))
print("Chéo chính tăng:", cheo_chinh_tang(mt))
print("Tam giác dưới chéo phụ:", tam_giac_duoi_cheo_phu(mt))
print("Chéo phụ giảm:", cheo_phu_giam(mt))
```

```

# 5. Pandas và Matplotlib
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Tháng": [1, 2, 3, 4, 5],
    "Doanh thu": [100, 150, 120, 180, 200]
})

print(df)

plt.plot(df["Tháng"], df["Doanh thu"])
plt.xlabel("Tháng")
plt.ylabel("Doanh thu")
plt.title("Biểu đồ doanh thu")
plt.grid(True)
plt.show()
```
